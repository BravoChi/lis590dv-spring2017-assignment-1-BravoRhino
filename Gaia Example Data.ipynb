# Downloading Gaia data

This notebook is a simple example of downloading Gaia data, loading it in, and doing simple plots of it.

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = {}
with h5py.File("gaia_validp.h5") as f:
#     print(f)
    for k in f:
        print(k)
        data[k] = f[k][:]
```

```python
print(data.keys())
```

```python
print(data)
```

```python
#Firstly, I would like to obtain lengths and distributions of each array
```

```python
print(len(data['pmra']))
plt.plot(data['pmra'], 'co')
plt.show()
```

```python
print(len(data['ra']))
plt.plot(data['ra'], 'bo')
plt.show()
```

```python
print(len(data['pmdec']))
plt.plot(data['pmdec'], 'yo')
plt.show()
```

```python
print(len(data['parallax']))
plt.plot(data['parallax'], 'go')
plt.show()
```

```python
print(len(data['phot_g_mean_mag']))
plt.plot(data['phot_g_mean_mag'], 'ro')
plt.show()
```

```python
print(len(data['dec']))
plt.plot(data['pmdec'], 'ko')
plt.show()
```

```python
#As we can see in those six graphs, the distributions are quite different from each other. 
#So, I think it is meaningless to combine these six arraies together.
```

```python
#SO I would like to analyze data set individually. And take 'dec' set as an example.
#As we can see in the above graph, the data is mainly concentrated in the range from -1000 to 1000.
#Therefore, I use a histogram to describe its detail distribution.
plt.hist(data['dec'],bins = 50, facecolor = 'black',edgecolor = 'white')
plt.show()
```
